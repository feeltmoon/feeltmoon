```python
import pandas as pd
import numpy as np

df = pd.read_excel(r'G:\WFH\20211226\ECS_spec.xlsx', sheet_name='Sheet1')
df1 = pd.read_excel(r'G:\WFH\20211226\BGB-16673-102_V1.0.xlsx', sheet_name='Fields')
```

```python
pd.set_option('display.max_rows', 3)
pd.set_option('display.max_columns', 85)
```

```python
df = pd.DataFrame(df,columns=['Version#','Serial #','CheckName','Folder OID','Form Name','Form OID','Field Label',
                               'Field OID','Check Type','Assigner','Programmed by'])
```

```python
df = df.rename(columns={"Field OID": "FieldOID"})
```

```python
df['FieldOID_spec'] = df['FieldOID']
```

```python
df1 = pd.DataFrame(df1,columns=['Ordinal','FormOID','PreText','FieldOID','DraftFieldActive','IsLog','IsVisible',
                               'DraftFieldNumber','DraftFieldName','VariableOID','DataFormat','DefaultValue','DataDictionaryName',
                               'UnitDictionaryName','CodingDictionary','ControlType','SASLabel','IsRequired',
                               'QueryNonConformance','QueryFutureDate', 'DoesNotBreakSignature',
                               'ViewRestrictions','EntryRestrictions','SourceDocument','ReviewGroups'])
```

```python
df1 = df1[~df1['Ordinal'].isna()]
df1 = df1[df1['DraftFieldActive'] == True]
```

```python
df1['FieldOID_als'] = df1['FieldOID']
```

```python
df1 = df1[df1['IsRequired'] == True]
```

```python
join = pd.merge(df1, df, how='outer', on=['FieldOID'])
```

```python
join['IsEqual'] = (join['FieldOID_spec']==join['FieldOID_als'])
join
```

```python
join = pd.DataFrame(join,columns=['Ordinal','FormOID','PreText','FieldOID','DraftFieldActive','IsLog','IsVisible',
                                  'DraftFieldNumber','DraftFieldName','VariableOID','DataFormat','DefaultValue','DataDictionaryName',
                                  'UnitDictionaryName','CodingDictionary','ControlType','SASLabel','IsRequired',
                                  'QueryNonConformance','QueryFutureDate', 'DoesNotBreakSignature',
                                  'ViewRestrictions','EntryRestrictions','SourceDocument','ReviewGroups',
                                  'Assigner','Form OID','IsEqual'])
```
